<a id="créer-une-colonne-ndate-qui-signifie-date-de-naissance-du-candidatqui-concatène-le-jour,-le-mois-et-l’année-de-naissance-au-format-jj/mm/aaaa"></a>
# Créer une colonne Ndate qui signifie date de naissance du candidatqui concatène le jour, le mois et l’année de naissance au format jj/mm/aaaa

```python
# S'assurer d'avoir extrait les composantes jour, mois, année
a = base["Matricule"].str[5:7]
b= base["Matricule"].str[7:9]
c= base["Matricule"].str[9:13]

# Créer la colonne Ndate au format jj/mm/aaaa
base["Ndate"] = (
    a + "/" +
    b+ "/" +
    c
)
```

<a id="créer-une-colonne-age-qui-calcule-l’age-des-candidats-à-ce-concours-en-2025."></a>
# Créer une colonne Age qui calcule l’age des candidats à ce concours en 2025.
 

Avant d’ajouter la colonne "Âge", nous nous sommes assurés que la colonne "Ndate" est bien au format date, afin de permettre un calcul correct de l’âge.

```python
# S'assurer que la colonne Ndate est bien au format date
base["Ndate"] = pd.to_datetime(base["Ndate"], format="%d/%m/%Y", errors="coerce")

# Calculer l'âge en 2025
base["Age"] = 2025 - base["Ndate"].dt.year
```

<a id="créer-une-colonne-moyenne-qui-calcule-la-moyenne-d’admission-de-chaque-étudiant"></a>
# Créer une colonne Moyenne qui calcule la moyenne d’admission de chaque étudiant

```python
# Création de la moyenne pondérée 
base['Moyenne'] = (
    (base['Maths'] * 4.5) +
    (base['Physique'] * 4.5) +
    (base['Anglais'] * 3) +
    (base['Français'] * 2)
) / 14.0

# Arrondir à 2 décimales
base['Moyenne'] = base['Moyenne'].round(2)
```

<a id="créer-la-colonne-rang-qui-sera-affecté-de-la-concaténation-«er»-si-le-premier-au-concours-est-un-homme-c’est-à-dire-1er-sinon-si-le-premier-au-concours-est-une-femme-on-affectera-la-concaténation-«ère»-c’est-à-dire-1ère;-pour-les-autres-on-ajoutera-simplement-«e»"></a>
# Créer la colonne Rang qui sera affecté de la concaténation «er» si le premier au concours est un homme c’est-à-dire 1er sinon si le premier au concours est une femme on affectera la concaténation «ère» c’est-à-dire 1ère; pour les autres on ajoutera simplement «e»

Pour le calcul du rang nous allons d’abord créer la colonne "Sexe" afin de pouvoir adapter dynamiquement l’affichage du rôle : "1er" pour un homme, et "1ère" pour une femme.

```python
# Créer la colonne Sexe à partir du Matricule
base['Sexe'] = base['Matricule'].apply(lambda x: 'Homme' if 'H' in x else 'Femme')

# Trier par Moyenne décroissante
base = base.sort_values(by='Moyenne', ascending=False).reset_index(drop=True)

# Créer une colonne Rang numérique
base['Rang_num'] = base.index + 1  # commence à 1

# Récupérer le sexe du premier classé
premier_sexe = base.loc[0, 'Sexe']
premier_suffixe = 'er' if premier_sexe == 'Homme' else 'ère'

# Créer la colonne Rang avec "e" par défaut
base['Rang'] = base['Rang_num'].astype(str) + 'e'

# Corriger le premier rang (1er ou 1ère)
base.loc[0, 'Rang'] = f"1{premier_suffixe}"

# Supprimer la colonne temporaire Rang_num
base.drop(columns='Rang_num', inplace=True)
```

<a id="créer-la-colonne-décision,-constituée-des-modalités-suivantes-admis-ou-echoué.-sont-admis-aux-concours-ceux-qui-ont-16-de-moyennes-et-plus,-les-autres-ne-vérifions-pas-cette-condition-sont-échoués."></a>
# Créer la colonne Décision, constituée des modalités suivantes: Admis ou Echoué. Sont admis aux concours ceux qui ont 16 de moyennes et plus, les autres ne vérifions pas cette condition sont échoués.

```python
# Création de la colonne Décision selon la moyenne
base['Décision'] = base['Moyenne'].apply(lambda x: 'Admis' if x >= 16 else 'Echoué')
```

# Construire une colonne Région

```python
# Définir les groupes de pays par région
pays_centre = ['Cameroun', 'Congo', 'Gabon', 'Tchad', 'RCA', 'Guinée Équatoriale', 'RDC', 'Angola']
pays_sud = ['Madagascar']
pays_est = ['Soudan']

# Fonction pour affecter la région
def attribuer_region(pays):
    if pays in pays_centre:
        return 'Centre'
    elif pays in pays_sud:
        return 'Sud'
    elif pays in pays_est:
        return 'Est'
    else:
        return 'Ouest'

# Appliquer la fonction à chaque ligne
base['Région'] = base['Pays'].apply(attribuer_region)
```

<a id="réaliser-une-analyse-de-correspondance-pour-vérifier-cette-assertion-et-conclure-par-un-test-statistique."></a>
# Réaliser une analyse de correspondance pour vérifier cette assertion et conclure par un test statistique.

Pour vérifier l’assertion selon laquelle les pays de l'Afrique de l'ouest ont des meilleurs performances au Bac par rapport à ceux de l'Afrique centrale nous allons faire une analyse factorielle des correspondances + test du Khi² après catégorisation des moyennes pour confirmer cette assertion. Pour cela nous allons commencer par créer la colonne Mention au Bac.

```python
def categoriser_moyenne(x):
    if x < 10:
        return 'Faible'
    elif x < 12:
        return 'Passable'
    elif x < 14:
        return 'Assez bien'
    elif x < 16:
        return 'Bien'
    else:
        return 'Très bien'

base['Mention au Bac'] = base['Moyenne au Bac'].apply(categoriser_moyenne)

# Créer la table de contingence
contingence = pd.crosstab(base['Région'], base['Mention au Bac'])
print("Table de contingence :\n", contingence)

# AFC
ca = prince.CA(n_components=2)
ca = ca.fit(contingence)

# Extraire les coordonnées
rows = ca.row_coordinates(contingence)
cols = ca.column_coordinates(contingence)

# Tracer le graphique manuellement
fig, ax = plt.subplots(figsize=(8,6))

# Modalités lignes (régions)
ax.scatter(rows[0], rows[1], color='blue', label='Régions')
for i, txt in enumerate(rows.index):
    ax.text(rows.iloc[i, 0], rows.iloc[i, 1], txt, color='blue')

# Modalités colonnes (mentions)
ax.scatter(cols[0], cols[1], color='red', label='Mentions')
for i, txt in enumerate(cols.index):
    ax.text(cols.iloc[i, 0], cols.iloc[i, 1], txt, color='red')

# Mise en forme du graphique
ax.set_title("AFC : Région vs Mention au Bac")
ax.set_xlabel("Axe 1")
ax.set_ylabel("Axe 2")
ax.axhline(0, color='grey', lw=1)
ax.axvline(0, color='grey', lw=1)
plt.grid()
plt.legend()
plt.show()
```

Après l’AFC, une tendance visuelle suggérait une supériorité des performances des candidats d’Afrique de l’Ouest par rapport à ceux d’Afrique centrale. Pour confirmer cette observation, nous avons réalisé un test du khi-deux d’indépendance entre la région (Ouest et Centre) et la mention au Bac.
Hypothèses:
H0: Il n’y a pas de lien entre la région et la mention au Bac.
H1: Il y a un lien significatif entre la région et la mention.

```python
# Filtrage des données pour ne garder que les régions Ouest et Centre
base2 = base[base["Région"].isin(["Ouest", "Centre"])]

# Création du tableau croisé
tableau = pd.crosstab(base2["Région"], base2["Mention au Bac"])

# Application du test du khi-deux
chi2, p ,_, expected = chi2_contingency(tableau)

# Calcul du V de Cramér
n = tableau.to_numpy().sum()  
phi2 = chi2 / n
k = min(tableau.shape)  
v_cramer = np.sqrt(phi2 / (k - 1))

# Affichage des résultats avec interprétation
print("Résultat du test du khi-deux ")
print(f"Statistique de test : {chi2:.4f}")
print(f"p-value : {p:.4f}")
print(f"V de Cramer : {v_cramer:.4f}")


alpha = 0.05
if p < alpha:
    print("L’hypothèse selon laquelle les candidats de l’Ouest ont de meilleures performances au Bac que ceux du Centre est appuyée par les données.")
else:
    print("L’hypothèse sur de meilleures performances des candidats de l’Ouest n’est pas confirmée.")
```

# Peut-on dire que la performance des candidats au Bac est liée au Pays ?

Visualisation des moyennes au bac par pays

```python
# Calcul de la moyenne au Bac par pays
moyennes_pays = base.groupby('Pays')['Moyenne au Bac'].mean().reset_index()

# Trier les pays par moyenne décroissante       
moyennes_pays = moyennes_pays.sort_values('Moyenne au Bac', ascending=False)

# Taille de la figure
plt.figure(figsize=(16, 8))

# Création du barplot
sns.barplot(x='Pays', y='Moyenne au Bac', data=moyennes_pays, palette='Set2')

# Affichage de la valeur de la moyenne au-dessus de chaque barre
for index, row in moyennes_pays.iterrows():
    plt.text(index, row['Moyenne au Bac'] + 0.1, f"{row['Moyenne au Bac']:.2f}", 
             ha='center', va='bottom', fontsize=9, rotation=90)

# Personnalisation du graphique
plt.xticks(rotation=45, ha='right')
plt.title("Moyenne au Bac par pays", fontsize=14)
plt.xlabel("Pays")
plt.ylabel("Moyenne au Bac")
plt.tight_layout()

# Affichage
plt.show()
```

Pour ce faire nous allons avoir deux approches, une qualitative (khi-deux plus V de Cramer) et une autre quantitative (ANOVA) 

Approche qualitative (khi-deux plus V de Cramer)

```python
# Crée un tableau de contingence entre 'Pays' et 'Mention au Bac'
contingence_table = pd.crosstab(base['Pays'], base['Mention au Bac'])

# Effectue le test du khi-deux
chi2, p_value, _, expected = chi2_contingency(contingence_table)

print(f"Statistique du khi-deux : {chi2:.4f}")
print(f"p-value : {p_value:.4f}")

if p_value < 0.05:
    print("On rejette H0 : il existe un lien significatif entre le pays et la mention au bac.")
else:
    print("On ne rejette pas H0 : aucun lien significatif entre le pays et la mention au bac.")

# Calcul du V de Cramer
n = contingence_table.sum().sum() 
min_dim = min(contingence_table.shape) - 1  
v_cramer = np.sqrt(chi2 / (n * min_dim))

print(f"V de Cramer : {v_cramer:.4f}")
```

Approche qualitative (ANOVA)

Nous allons d’abord commencé par vérifier si nos données suivent une distribution normale ou non. Afin d’améliorer notre analyse, nous allons réalisé un test statistique, le test de Jarque-Bera, qui permet d’évaluer la normalité des données. Nous avons choisi ce test car il est particulièrement adapté à la taille importante de notre base de données.
H0: Les données suivent une loi normale
H1: Les données ne suivent pas une loi normale

```python
def tester_normalite(df):
    colonnes_numeriques = df.select_dtypes(include=[np.number]).columns

    for variable in colonnes_numeriques:
        donnees = df[variable].dropna()

        print(f"\n Variable : {variable} ")

        
        # Test de Jarque-Bera
        stat_jb, pval_jb = jarque_bera(donnees)
        print(f"Statistique Jarque-Bera = {stat_jb:.4f}, p-valeur = {pval_jb:.4f}")
        if pval_jb > 0.05:
            print(" Données semblent suivre une loi normale ")
        else:
            print(" Données ne suivent pas une loi normale ")

        # Visualisation
        plt.figure(figsize=(12, 5))
        plt.subplot(1, 2, 1)
        sns.histplot(donnees, kde=True, bins=30, color='lightblue')
        plt.title(f'Histogramme et densité de {variable}')
        plt.subplot(1, 2, 2)
        probplot(donnees, dist="norm", plot=plt)
        plt.title(f'QQ-plot de {variable}')
        plt.tight_layout()
        plt.show()


tester_normalite(base)
```

Vu que les données ne suivent pas une loi normale et ne sont pas homoscédastiques alors nos faisant le test de kruskal-Wallis 

```python
# Création des groupes de moyennes au Bac par pays
groupes = [base[base['Pays'] == pays]['Moyenne au Bac'] for pays in base['Pays'].unique()]

# Application du test de Kruskal-Wallis
stat, p_value = kruskal(*groupes)

# Résultats
print("Test de Kruskal-Wallis : Influence du pays sur la moyenne au Bac")
print(f"Statistique : {stat:.4f}")
print(f"p-value : {p_value:.4f}")

# Interprétation
alpha = 0.05
if p_value < alpha:
    print("Il existe une différence significative de la moyenne au Bac selon le pays d'origine des candidats.")
else:
    print("Aucune différence significative n’a été détectée entre les pays pour la moyenne au Bac.")
```

test de Hosmer-Lemeshow

```python

def hosmer_lemeshow_test(y_true, y_prob, g=10):
    # Créer un DataFrame avec les vrais labels et les probabilités prédites
    df = pd.DataFrame({'y_true': y_true, 'y_prob': y_prob})

    # Créer des groupes basés sur les quantiles des probabilités
    df['quantile'] = pd.qcut(df['y_prob'], q=g, duplicates='drop')

    # Calcul des événements et non-événements observés
    obs = df.groupby('quantile')['y_true'].agg(['sum', 'count']).rename(columns={'sum': 'observed_events', 'count': 'total'})
    obs['observed_nonevents'] = obs['total'] - obs['observed_events']

    # Calcul des événements attendus
    exp = df.groupby('quantile')['y_prob'].agg(['sum']).rename(columns={'sum': 'expected_events'})
    exp['expected_nonevents'] = obs['total'] - exp['expected_events']

    # Fusionner les observations et les attentes
    result = pd.concat([obs, exp], axis=1)

    # Calcul de la statistique du test
    chi2_stat = (
        ((result['observed_events'] - result['expected_events']) ** 2) / result['expected_events'] +
        ((result['observed_nonevents'] - result['expected_nonevents']) ** 2) / result['expected_nonevents']
    ).sum()

    # Degrés de liberté
    dof = len(result) - 2
    p_value = 1 - chi2_dist.cdf(chi2_stat, dof)

    return chi2_stat, p_value


hl_stat, hl_pvalue = hosmer_lemeshow_test(y, y_proba)

print(f"\nStatistique Hosmer-Lemeshow : {hl_stat:.4f}")
print(f"p-value : {hl_pvalue:.4f}")
if hl_pvalue > 0.05:
    print("Le modèle est bien calibré (p > 0.05)")
else:
    print("Le modèle n'est pas bien calibré (p <= 0.05)")
```
